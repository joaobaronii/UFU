1. Uma caixa contém duas bolas, que podem ser ambas vermelhas, ambas azuis, ou uma de cada cor. A cada passo, uma bola é sorteada e substituída por uma nova bola, que tem 80% de chance de ser da mesma cor. Se inicialmente ambas são vermelhas, qual a probabilidade de que a quinta bola retirada seja vermelha?

```python
import random

def mc_caixa(num_simulacoes):
    sucesso_quinta_vermelha = 0
    
    for _ in range(num_simulacoes):
        caixa = ['R', 'R']

        for _ in range(4):
            bola_sorteada = random.choice(caixa)

            if random.random() < 0.80:
                nova_bola = bola_sorteada 
            else:
                if bola_sorteada == 'R':
                    nova_bola = 'B'
                else:
                    nova_bola = 'R'

            caixa.remove(bola_sorteada)
            caixa.append(nova_bola)
            
        quinta_bola = random.choice(caixa)
        
        if quinta_bola == 'R':
            sucesso_quinta_vermelha += 1
            
    return sucesso_quinta_vermelha / num_simulacoes


n = 100000
resultado = mc_caixa(n)

print(f"Resultado após {n} simulações: {resultado:.4f}")
```

2. Numa estrada há carros e caminhões. Três de cada quatro caminhões na estrada são seguidos por um carro, enquanto que apenas um de cada cinco carros é seguido por um caminhão. Qual a porcentagem de caminhões na estrada?

```python
def mc_caminhao(qtd_simulacoes):
    estado_atual = random.choice(['T', 'C'])
    caminhoes = 0

    for _ in range(qtd_simulacoes):
        if estado_atual == 'T':
            caminhoes += 1
            if random.random() < 0.75:
                estado_atual = 'C'
            else:
                estado_atual = 'T'
        else:
            if random.random() < 0.20:
                estado_atual = 'T'
            else:
                estado_atual = 'C'

    return caminhoes / qtd_simulacoes


resultado = mc_caminhao(n)

print(f"Resultado após {n} simulações: {resultado:.4f}")
```

```python
def mc_defeituoso(num_simulacoes):
    um_defeito = 0
    prob_defeito = 0.001
    
    for _ in range(num_simulacoes):
        defeito = []
        for i in range(3):
            sorteio = random.random()
            if sorteio < prob_defeito:
                defeito.append(True)
            else:
                defeito.append(False)
        
        if any(defeito):
            um_defeito += 1
            
    return um_defeito / num_simulacoes


resultado = mc_defeituoso(n)

print(f"Resultado após {n} simulações: {resultado:.4f}")
```
