1. Helena está na janela de casa observando uma tempestade de raios no horizonte. Um raio cai em média a cada 45 segundos. Se Helena não vê nenhum raio no primeiro minuto, ela cansa de esperar e sai da janela. Senão, ela observa por mais dois minutos.

    (a) Qual o número esperado de raios que Helena observa?

    (b) Qual o tempo esperado que ela fica na janela?  

```python
import numpy as np

def mc_raios(n_simulacoes):
    raios_por_segundo = 1/45
    n_raios = 0
    tempo_total = 0
    
    for _ in range(n_simulacoes):
        raios_primeiro_minuto = np.random.poisson(raios_por_segundo * 60)
        
        if raios_primeiro_minuto == 0:
            tempo = 60
            raios = 0
        else:
            raios_depois = np.random.poisson(raios_por_segundo * 120)
            
            tempo = 60 + 120
            raios = raios_primeiro_minuto + raios_depois
            
        n_raios += raios
        tempo_total += tempo

    raios_esperados = n_raios / n_simulacoes
    tempo_esperado = tempo_total / n_simulacoes
    
    return raios_esperados, tempo_esperado


raios, tempo = mc_raios(100000)

print(f"Número esperado de raios: {raios:.4f}")
print(f"Tempo esperado na janela: {tempo:.4f}")
```

2. Cem itens são simultaneamente colocados num teste de vida. Suponha que os tempos de vida de cada um dos itens são variáveis independentes e que seguem uma distribuição exponencial com média 200 horas. O teste termina quando há um total de 5 falhas. Qual o tempo esperado da duração do teste? 

```python
import numpy as np

def mc_teste_vida(n_simulations):
    n_itens = 100        
    n_falhas = 5   
    vida_media = 200  
    
    tempos_vida = np.random.exponential(scale=vida_media, size=(n_simulations, n_itens))
    
    tempos_vida.sort(axis=1)
    
    duracao_testes = tempos_vida[:, n_falhas-1]
    
    tempo_esperado = np.mean(duracao_testes)
    
    return tempo_esperado


tempo = mc_teste_vida(100000)

print(f"Tempo esperado da duração do teste: {tempo:.4f} horas")
```
